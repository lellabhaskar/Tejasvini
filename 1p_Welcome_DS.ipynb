```python
class FiboGen:
  def __iter__(self):
    self.n1=-1
    self.n2=1
    self.n3=self.n1+self.n2
    return self
  def __next__(self):
    self.n1=self.n2
    self.n2=self.n3
    self.n3=self.n1+self.n2
    return self.n3
prop =FiboGen()
pointer_to_prob=iter(prop)
#print(next(pointer_to_prob))
#print(next(pointer_to_prob))
for i in range(5):
  print(next(pointer_to_prob))
```

```python
 # Diagnosis  => M and B
 #Encoding means converting data from one format to another format
diagnosis =data.loc[:,'diagnosis']
diagnosis   #diagnosis[:20]
```

```python
# find the count of M and B in diagnosis Column
print(len(data[data.diagnosis=='M']))
print(len(data[data.diagnosis=='B']))
```

```python
#Color Encoding
#Malign -> Red, Benign -> Blue
encoding_logic=lambda val: 'red' if val=='M' else 'blue'
diagnosis_encoded =diagnosis.map(encoding_logic)
diagnosis_encoded
```
