```python
class FiboGen:
  def __iter__(self):
    self.n1=-1
    self.n2=1
    self.n3=self.n1+self.n2
    return self
  def __next__(self):
    self.n1=self.n2
    self.n2=self.n3
    self.n3=self.n1+self.n2
    return self.n3
prop =FiboGen()
pointer_to_prob=iter(prop)
#print(next(pointer_to_prob))
#print(next(pointer_to_prob))
for i in range(5):
  print(next(pointer_to_prob))
```
